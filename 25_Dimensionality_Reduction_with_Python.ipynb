```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Memuat dataset Iris (4 dimensi: sepal length, sepal width, petal length, petal width)
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Standarisasi data (Wajib untuk PCA agar skala fitur sama)
X_scaled = StandardScaler().fit_transform(X)

print(f"Bentuk data asli: {X.shape}")
```

```python
# Menerapkan PCA ke 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualisasi PCA
plt.figure(figsize=(8, 6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_name)

plt.title('Visualisasi PCA - Dataset Iris (Reduksi ke 2D)')
plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]:.2%})')
plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]:.2%})')
plt.legend()
plt.grid(True)
plt.show()

# Penjelasan Makna Hasil PCA
print(f"Total varians yang dipertahankan oleh PCA: {np.sum(pca.explained_variance_ratio_):.2%}")
```

```python
# Menerapkan t-SNE ke 2D
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)

# Visualisasi t-SNE
plt.figure(figsize=(8, 6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=target_name)

plt.title('Visualisasi t-SNE - Dataset Iris (Reduksi ke 2D)')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
